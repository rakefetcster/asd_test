```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
asd_data = pd.read_csv('data/autism_screening.csv')
asd_data

```

```python
#Arranging the table, adding data and replacing the ASD column with 0/1 instead of true/false 
asd_data['age'].fillna(30.0, inplace=True)
asd_data['Class/ASD'] = asd_data['Class/ASD'].replace({'YES': 1, 'NO': 0})
asd_data
```

```python
#We will look for a question that best predicts autistic people
asd_data.corr(numeric_only=True)
```

```python
# Questions 6 and 9 best predict autism, 0 - not diagnosed with autism, 1 - diagnosed with autism 
asd_data.groupby(['Class/ASD'])[['A6_Score','A9_Score']].mean().plot(kind='bar')
```
